### Class methods

```python
class student:

    def __init__ (self,name ,age,gpa):
        self.name = name 
        self.age  = age
        self.gpa = gpa
    
    def show_detail(self):
        print("____ student_data ____")
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"GPA : {self.gpa}")

s1= student("shakoor",20,3.21)
s1.show_detail()

```
