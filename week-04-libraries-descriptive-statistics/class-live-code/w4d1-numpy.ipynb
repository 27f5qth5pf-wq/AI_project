```python
import time
```

```python
n = 100000

python_list = list(range(0, n))

start = time.time()

python_list = [i+5 for i in python_list]

end = time.time()

print("start : ", start)
print("end : ", end)
```

```python
(end - start) * 1000
```

```python
import numpy as np
```

```python
numpy_arr = np.arange(0, n)

start = time.time()

numpy_arr = numpy_arr + 5

end = time.time()

print("Numpy time : ", (end - start)*1000)
```

```python
numpy_arr.itemsize * numpy_arr.size
```
