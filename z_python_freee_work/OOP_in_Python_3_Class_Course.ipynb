## 1.1 Why do we need OOP?

So far, you have probably written programs using **functions and variables** (this is called *procedural programming*). That works fine for small programs, but as programs grow, data and the functions that work on that data get scattered everywhere.

**The problem — procedural style:**

```python
# Procedural approach: data and functions are SEPARATE
student1_name = "Ali"
student1_marks = [85, 90, 78]

student2_name = "Sara"
student2_marks = [92, 88, 95]

def calculate_average(marks):
    return sum(marks) / len(marks)

print(student1_name, "->", calculate_average(student1_marks))
print(student2_name, "->", calculate_average(student2_marks))

# Imagine 500 students... 1000 variables! Messy and error-prone.
```

**The OOP solution:** bundle the *data* (name, marks) and the *behavior* (calculate average) together into a single unit called a **class**. Each student becomes an **object** created from that class.

> 💡 **Analogy:** A class is like a **blueprint of a house**. The blueprint itself is not a house — but from one blueprint you can build many actual houses (objects), each with its own paint color, furniture, and residents.

**The 4 pillars of OOP** (we will cover all of them in this course):
1. **Encapsulation** – bundling data + methods, and protecting the data (Class 3)
2. **Inheritance** – creating new classes from existing ones (Class 2)
3. **Polymorphism** – same interface, different behavior (Class 2)
4. **Abstraction** – hiding complex details, exposing only what's needed (Class 3)

## 1.2 Your first class and object

- `class` keyword defines a class. Class names use **PascalCase** by convention (`Student`, `BankAccount`).
- Creating an object from a class is called **instantiation**.

```python
# The simplest possible class
class Dog:
    pass   # 'pass' means "empty for now"

# Creating objects (instances) of the class
dog1 = Dog()
dog2 = Dog()

print(type(dog1))          # <class '__main__.Dog'>
print(dog1 is dog2)        # False -> two SEPARATE objects in memory
```

## 1.3 The `__init__` method and `self`

Real objects need **data**. The `__init__` method (the **constructor**) runs *automatically* every time you create an object. It is where we set up the object's initial data.

- `self` refers to **the specific object being created/used**. Python passes it automatically.
- Variables attached to `self` (like `self.name`) are called **instance attributes** — every object gets its own copy.

```python
class Student:
    def __init__(self, name, age, grade):
        # 'self' = the new object being created
        self.name = name      # instance attribute
        self.age = age
        self.grade = grade
        print(f"__init__ was called! Created student: {name}")

# When we write Student("Ali", 16, "10th"), Python calls:
#   __init__(self=<new object>, name="Ali", age=16, grade="10th")
s1 = Student("Ali", 16, "10th")
s2 = Student("Sara", 17, "11th")

print(s1.name, "-", s1.age, "-", s1.grade)   # access attributes with dot notation
print(s2.name, "-", s2.age, "-", s2.grade)
```

> ⚠️ **Common beginner mistakes:**
> 1. Forgetting `self` as the first parameter of a method → `TypeError: ... takes 0 positional arguments but 1 was given`
> 2. Writing `name = name` instead of `self.name = name` inside `__init__` → the data is lost when the method ends.
> 3. Calling the class without required arguments → `Student()` fails because `__init__` expects name, age, grade.

## 1.4 Methods — giving objects behavior

A **method** is simply a function defined *inside* a class. It always takes `self` first, which gives it access to that object's own data.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks          # a list of numbers

    def average(self):              # method (no extra data needed)
        return sum(self.marks) / len(self.marks)

    def add_mark(self, mark):       # method that takes an argument
        self.marks.append(mark)

    def has_passed(self):           # methods can use other methods
        return self.average() >= 50

s1 = Student("Ali", [85, 90, 78])
s2 = Student("Sara", [40, 45, 50])

print(f"{s1.name}'s average: {s1.average():.1f}  | Passed? {s1.has_passed()}")
print(f"{s2.name}'s average: {s2.average():.1f}  | Passed? {s2.has_passed()}")

s2.add_mark(95)   # Sara's data changes; Ali's is untouched
print(f"{s2.name}'s new average: {s2.average():.1f} | Passed? {s2.has_passed()}")
```

Notice the beauty: `s1.average()` and `s2.average()` run the **same code** but give **different answers**, because `self` points to a different object each time.

## 1.5 Instance attributes vs class attributes

| | Instance attribute | Class attribute |
|---|---|---|
| Defined | inside `__init__` with `self.` | directly inside the class body |
| Belongs to | **each object separately** | **the class itself (shared by all)** |
| Use for | data unique to each object | constants / shared counters |

```python
class Student:
    school = "City Public School"   # CLASS attribute -> shared by everyone
    count = 0                       # shared counter

    def __init__(self, name):
        self.name = name            # INSTANCE attribute -> unique per object
        Student.count += 1          # update the shared counter

s1 = Student("Ali")
s2 = Student("Sara")
s3 = Student("Bilal")

print(s1.school, "|", s2.school)        # same shared value
print("Total students created:", Student.count)

# Changing the class attribute affects ALL objects
Student.school = "Model High School"
print(s1.school, "|", s3.school)
```

## 1.6 Printing objects nicely: `__str__`

By default, printing an object shows something ugly like `<__main__.Student object at 0x7f...>`. Define `__str__` to control what `print()` shows.

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def __str__(self):
        return f"Student({self.name}, grade {self.grade})"

s = Student("Ali", "10th")
print(s)              # Student(Ali, grade 10th)  <- __str__ is used automatically
```

## 1.7 Worked example — putting it all together: `BankAccount`

Read this class top to bottom, then run it. This is the pattern you will use in the practice session.

```python
class BankAccount:
    bank_name = "Python National Bank"        # class attribute

    def __init__(self, owner, balance=0):     # default argument
        self.owner = owner
        self.balance = balance
        self.transactions = []                # every account gets its own history

    def deposit(self, amount):
        if amount <= 0:
            print("❌ Deposit must be positive.")
            return
        self.balance += amount
        self.transactions.append(f"Deposited {amount}")
        print(f"✅ {self.owner} deposited {amount}. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"❌ Insufficient funds! Balance is {self.balance}")
            return
        self.balance -= amount
        self.transactions.append(f"Withdrew {amount}")
        print(f"✅ {self.owner} withdrew {amount}. New balance: {self.balance}")

    def show_history(self):
        print(f"--- {self.owner}'s transactions ---")
        for t in self.transactions:
            print("  •", t)

    def __str__(self):
        return f"[{BankAccount.bank_name}] {self.owner}: Rs. {self.balance}"


acc = BankAccount("Ahmed", 1000)
acc.deposit(500)
acc.withdraw(300)
acc.withdraw(5000)        # should fail safely
acc.show_history()
print(acc)
```

```python
# Your code here

class counter:

    def __init__(self , count=0):
        self.count = count

    def increment(self):
        self.count += 1
        return self.count

    def decrement(self):
        if self.count <= 0:  
            print("Counter cannot go below zero.")
        else: 
            self.count -= 1
            return self.count

    def reset(self):
        self.count = 0
        return self.count
```

Create a class `ShoppingCart`:
- `add_item(name, price)` stores items in a dictionary `self.items`
- `remove_item(name)` removes an item (handle the case where it doesn't exist)
- `total()` returns the sum of all prices
- `__str__` shows all items and the total, nicely formatted

```python
# Your code here
class ShoppingCart:
    def __init__(self):
        self.items = {}

    def add_item(self, name, price):
        self.items[name] = price
        print(f"Added '{name}' (${price:.2f}) to the cart.")

    def remove_item(self, name):
        if name in self.items:
            del self.items[name]
            print(f"Removed '{name}' from the cart.")
        else:
            print(f"Could not remove: '{name}' is not in the cart.")

    def total(self):
        return sum(self.items.values())

    def __str__(self):
        if not self.items:
            return "Shopping Cart is empty."
    
        receipt = "--- Shopping Cart ---\n"
        for name, price in self.items.items():
            receipt += f"• {name}: ${price:.2f}\n"
        receipt += "---------------------\n"
        receipt += f"Total: ${self.total():.2f}"
        
        return receipt
```

## 2.1 What is inheritance?

**Inheritance** lets a new class (the **child** / subclass) reuse and extend the code of an existing class (the **parent** / superclass).

> 💡 **Analogy:** You inherit traits from your parents (eye color, height) — but you also have your own unique traits. A `Dog` *is an* `Animal`: it eats and sleeps like every animal, but it also barks.

**Why it matters:** Don't Repeat Yourself (**DRY**). Write common code once in the parent; every child gets it for free.

**Syntax:** `class Child(Parent):`

```python
class Animal:                        # PARENT class
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")


class Dog(Animal):                   # CHILD class — inherits everything from Animal
    def bark(self):                  # plus its own new method
        print(f"{self.name} says: Woof! 🐶")


class Cat(Animal):
    def meow(self):
        print(f"{self.name} says: Meow! 🐱")


d = Dog("Rex")
d.eat()      # inherited from Animal — we never wrote eat() in Dog!
d.sleep()    # inherited
d.bark()     # Dog's own method

c = Cat("Whiskers")
c.eat()
c.meow()
# d.meow()   # ❌ AttributeError — Dog didn't inherit Cat's methods
```

## 2.2 Method overriding

A child class can **replace** (override) a parent's method by defining a method with the **same name**. Python always uses the *most specific* version.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a generic sound.")


class Dog(Animal):
    def speak(self):                     # OVERRIDES Animal.speak
        print(f"{self.name} barks: Woof!")


class Cat(Animal):
    def speak(self):
        print(f"{self.name} meows: Meow!")


class Fish(Animal):
    pass                                  # no override -> uses parent's version


Dog("Rex").speak()        # child version
Cat("Whiskers").speak()   # child version
Fish("Nemo").speak()      # parent version (fallback)
```

## 2.3 `super()` — extending instead of replacing

Often you don't want to *throw away* the parent's method — you want to **run it AND add something extra**. `super()` calls the parent's version.

The most common use: a child's `__init__` that adds new attributes.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hi, I'm {self.name}, {self.age} years old.")


class Teacher(Person):
    def __init__(self, name, age, subject, salary):
        super().__init__(name, age)      # let the parent set name & age
        self.subject = subject           # then add the new attributes
        self.salary = salary

    def introduce(self):
        super().introduce()              # run parent's introduce first...
        print(f"   I teach {self.subject}.")   # ...then add more


class Student(Person):
    def __init__(self, name, age, grade):
        super().__init__(name, age)
        self.grade = grade


t = Teacher("Ms. Ayesha", 35, "Computer Science", 90000)
t.introduce()

s = Student("Ali", 16, "10th")
s.introduce()                            # inherited unchanged
```

> ⚠️ **Common mistake:** If a child class defines its own `__init__` but forgets `super().__init__(...)`, the parent's attributes are **never created** → `AttributeError` later.

## 2.4 Multilevel and multiple inheritance

- **Multilevel:** `A → B → C` (grandparent → parent → child). C inherits from *both* levels.
- **Multiple:** `class C(A, B)` — one child, two parents. Python resolves conflicts using the **MRO** (Method Resolution Order): left-to-right.

```python
# --- Multilevel inheritance ---
class Vehicle:
    def start(self):
        print("Engine started 🔑")

class Car(Vehicle):
    def drive(self):
        print("Driving on 4 wheels")

class SportsCar(Car):
    def turbo(self):
        print("TURBO MODE 🚀")

sc = SportsCar()
sc.start()    # from grandparent
sc.drive()    # from parent
sc.turbo()    # own method

# --- Multiple inheritance ---
class Camera:
    def take_photo(self):
        print("📸 Click!")

class Phone:
    def make_call(self):
        print("📞 Calling...")

class SmartPhone(Camera, Phone):     # inherits from BOTH
    pass

sp = SmartPhone()
sp.take_photo()
sp.make_call()

print(SmartPhone.__mro__)   # the order Python searches for methods
```

## 2.5 Polymorphism

**Polymorphism** = "many forms". The *same method call* behaves differently depending on the object's actual type. This is what makes overriding so powerful.

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14159 * self.r ** 2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base, self.height = base, height
    def area(self):
        return 0.5 * self.base * self.height


# ⭐ THE KEY IDEA: one loop, one method name, different behavior per object
shapes = [Rectangle(4, 5), Circle(3), Triangle(6, 8)]

for shape in shapes:
    print(f"{type(shape).__name__:<10} area = {shape.area():.2f}")
```

### Duck typing 🦆
Python doesn't even require a common parent — if an object *has* the method, Python will call it.
> *"If it walks like a duck and quacks like a duck, it's a duck."*

```python
class Duck:
    def sound(self):
        return "Quack!"

class Robot:                 # not related to Duck at all!
    def sound(self):
        return "Beep boop!"

def make_it_speak(thing):
    print(thing.sound())     # works on ANYTHING that has .sound()

make_it_speak(Duck())
make_it_speak(Robot())
```

## 2.6 `isinstance()` and `issubclass()`

Useful for checking types safely — note that a child object **is** an instance of its parent class too.

```python
class Animal: pass
class Dog(Animal): pass

d = Dog()
print(isinstance(d, Dog))        # True
print(isinstance(d, Animal))     # True  <- a Dog IS an Animal
print(isinstance(d, str))        # False
print(issubclass(Dog, Animal))   # True
print(issubclass(Animal, Dog))   # False (inheritance is one-way)
```

## 2.7 Worked example — Employee management with inheritance

```python
class Employee:
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def calculate_salary(self):
        return self.base_salary

    def __str__(self):
        return f"{type(self).__name__}: {self.name} → Rs. {self.calculate_salary():,}"


class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name, base_salary)
        self.bonus = bonus

    def calculate_salary(self):                    # override
        return self.base_salary + self.bonus


class SalesPerson(Employee):
    def __init__(self, name, base_salary, sales, commission_rate=0.05):
        super().__init__(name, base_salary)
        self.sales = sales
        self.commission_rate = commission_rate

    def calculate_salary(self):                    # override
        return int(self.base_salary + self.sales * self.commission_rate)


# Polymorphism in action: one list, one loop, correct salary for each type
staff = [
    Employee("Imran", 50000),
    Manager("Ayesha", 80000, bonus=20000),
    SalesPerson("Bilal", 40000, sales=500000),
]

total = 0
for emp in staff:
    print(emp)                       # __str__ calls the RIGHT calculate_salary
    total += emp.calculate_salary()
print(f"\n💰 Total payroll: Rs. {total:,}")
```

## 3.1 Encapsulation — protecting your data

**Encapsulation** = bundling data with the methods that operate on it, and **controlling access** to that data.

**The problem:** with plain attributes, anyone can put garbage in your object:

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

acc = BankAccount(1000)
acc.balance = -999999        # 😱 nothing stops this!
print(acc.balance)
```

**Python's access conventions:**

| Style | Example | Meaning |
|-------|---------|---------|
| public | `self.name` | Free to access from anywhere |
| protected (convention) | `self._name` | "Internal — please don't touch from outside" (not enforced) |
| private (name mangling) | `self.__name` | Python renames it to `_ClassName__name`, making outside access hard |

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner            # public
        self.__balance = balance      # private (double underscore)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return True
        print("❌ Invalid withdrawal.")
        return False

    def get_balance(self):            # controlled read access
        return self.__balance


acc = BankAccount("Ali", 1000)
acc.deposit(500)
print("Balance:", acc.get_balance())

# Direct access fails:
try:
    print(acc.__balance)
except AttributeError as e:
    print("⛔", e)

# (Technically reachable as acc._BankAccount__balance — Python trusts
#  programmers, but the convention says: DON'T.)
```

## 3.2 `@property` — the Pythonic getter/setter

Writing `get_x()` / `set_x()` everywhere is clunky. Python's `@property` lets you keep the *clean attribute syntax* (`obj.x`) while still running **validation code** behind the scenes.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age            # this line actually calls the setter below!

    @property
    def age(self):                # GETTER — runs when you READ obj.age
        return self.__age

    @age.setter
    def age(self, value):         # SETTER — runs when you WRITE obj.age = ...
        if not (5 <= value <= 120):
            raise ValueError(f"Age must be between 5 and 120, got {value}")
        self.__age = value


s = Student("Ali", 16)
print(s.age)          # looks like a normal attribute — getter runs

s.age = 17            # setter runs, validation passes
print(s.age)

try:
    s.age = -3        # setter runs, validation FAILS
except ValueError as e:
    print("⛔", e)
```

**Read-only properties** — a property with no setter cannot be changed from outside:

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def area(self):                      # computed on the fly, no parentheses needed
        return 3.14159 * self.radius ** 2

c = Circle(5)
print("Area:", c.area)      # NOT c.area() — it behaves like an attribute

try:
    c.area = 100            # ⛔ no setter defined
except AttributeError as e:
    print("⛔ Cannot set:", e)
```

## 3.3 Abstraction — abstract base classes (ABC)

**Abstraction** = defining *what* a class must do, without saying *how*. Think of it as a **contract**: "every payment method MUST have a `pay()` method — I don't care how each one implements it."

- Use the `abc` module: inherit from `ABC` and mark required methods with `@abstractmethod`.
- You **cannot create objects** of an abstract class — only of children that implement *all* abstract methods.

```python
from abc import ABC, abstractmethod

class PaymentMethod(ABC):                 # abstract class — the "contract"
    @abstractmethod
    def pay(self, amount):
        pass                              # no implementation here

class CreditCard(PaymentMethod):
    def pay(self, amount):
        print(f"💳 Paid Rs. {amount} by credit card")

class EasyPaisa(PaymentMethod):
    def pay(self, amount):
        print(f"📱 Paid Rs. {amount} via EasyPaisa")

class Cash(PaymentMethod):
    def pay(self, amount):
        print(f"💵 Paid Rs. {amount} in cash")


# ⛔ Cannot instantiate the abstract class itself:
try:
    p = PaymentMethod()
except TypeError as e:
    print("⛔", e)

# ✅ But children work — and are interchangeable (polymorphism!)
for method in [CreditCard(), EasyPaisa(), Cash()]:
    method.pay(2500)
```

> 💡 **Why bother?** In a big team, the abstract class *guarantees* that anyone adding a new payment method can't forget to write `pay()` — Python will refuse to create the object otherwise. Abstraction = fewer bugs, clearer design.

## 3.4 Magic (dunder) methods

Methods with **d**ouble **under**scores (`__init__`, `__str__`, ...) are called **dunder/magic methods**. Python calls them automatically in response to built-in operations. They let YOUR objects behave like built-in types.

| You write | Python secretly calls |
|-----------|----------------------|
| `print(obj)` | `obj.__str__()` |
| `obj` in a debugger / list | `obj.__repr__()` |
| `len(obj)` | `obj.__len__()` |
| `a == b` | `a.__eq__(b)` |
| `a + b` | `a.__add__(b)` |
| `a < b` | `a.__lt__(b)` |
| `obj[key]` | `obj.__getitem__(key)` |

```python
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add(self, song):
        self.songs.append(song)

    def __str__(self):                    # for print() — human-friendly
        return f"🎵 Playlist '{self.name}' ({len(self.songs)} songs)"

    def __repr__(self):                   # for debugging — precise
        return f"Playlist(name={self.name!r}, songs={self.songs!r})"

    def __len__(self):                    # enables len(playlist)
        return len(self.songs)

    def __getitem__(self, index):         # enables playlist[0] and for-loops!
        return self.songs[index]

    def __add__(self, other):             # enables playlist1 + playlist2
        combined = Playlist(f"{self.name} + {other.name}")
        combined.songs = self.songs + other.songs
        return combined


p1 = Playlist("Workout")
p1.add("Eye of the Tiger")
p1.add("Stronger")

p2 = Playlist("Chill")
p2.add("Weightless")

print(p1)                 # __str__
print(len(p1))            # __len__
print(p1[0])              # __getitem__

for song in p1:           # __getitem__ makes it iterable too!
    print("  ▶", song)

mega = p1 + p2            # __add__
print(mega)
print(repr(mega))         # __repr__
```

```python
# __eq__ and __lt__ : comparing objects YOUR way
class Money:
    def __init__(self, amount):
        self.amount = amount

    def __eq__(self, other):
        return self.amount == other.amount

    def __lt__(self, other):              # "less than" — also enables sorted()!
        return self.amount < other.amount

    def __repr__(self):
        return f"Rs.{self.amount}"

a, b, c = Money(500), Money(500), Money(200)
print(a == b)                  # True  (without __eq__ this would be False!)
print(c < a)                   # True
print(sorted([a, c, b]))       # sorting works because of __lt__
```

## 3.5 `@classmethod` and `@staticmethod`

| Type | First param | Use when... |
|------|------------|-------------|
| instance method | `self` | you need the object's data (most common) |
| `@classmethod` | `cls` | you need the class itself — e.g. alternative constructors |
| `@staticmethod` | none | it's just a related utility function |

```python
class Pizza:
    def __init__(self, size, toppings):
        self.size = size
        self.toppings = toppings

    @classmethod
    def margherita(cls, size):            # alternative constructor
        return cls(size, ["tomato", "mozzarella"])

    @classmethod
    def pepperoni(cls, size):
        return cls(size, ["tomato", "mozzarella", "pepperoni"])

    @staticmethod
    def inches_to_cm(inches):             # utility — needs no object data
        return inches * 2.54

    def __str__(self):
        return f"{self.size}\" pizza with {', '.join(self.toppings)}"

p1 = Pizza.margherita(12)                 # called on the CLASS, not an object
p2 = Pizza.pepperoni(16)
print(p1)
print(p2)
print("16 inches =", Pizza.inches_to_cm(16), "cm")
```

## 3.6 Composition — "has-a" vs "is-a"

Inheritance models **is-a** (a Dog *is an* Animal). **Composition** models **has-a** (a Car *has an* Engine) — you build objects out of other objects. Prefer composition when "is-a" doesn't naturally fit.

```python
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower

    def start(self):
        print(f"Engine ({self.horsepower} HP) started: vroom! 🔥")


class Car:                                # Car HAS-AN Engine (not "is an engine")
    def __init__(self, brand, horsepower):
        self.brand = brand
        self.engine = Engine(horsepower)  # composition!

    def start(self):
        print(f"{self.brand}: turning key...")
        self.engine.start()               # delegate to the part

my_car = Car("Toyota", 150)
my_car.start()
```
