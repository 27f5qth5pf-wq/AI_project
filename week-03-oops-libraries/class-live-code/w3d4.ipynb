```python
class Shape:

    def area(self):
        return 0
    def perimeter(self):
        0

class Rectangle(Shape):

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius



```

```python
class counter:
    def __init__(self,count=0):
        self.count = count

    def increment(self):
        self.count += 1
        return self.count

    def decrement(self):
        if self.count > 0:
            self.count -= 1
            return self.count
        else:
            print("Count is already zero. Cannot decrement further.")

    def reset(self):
        self.count = 0
    

```

```python
count1 = counter(2)
print(count1.count)
print(count1.increment())
print(count1.decrement())
print(count1.decrement())
print(count1.decrement())
print(count1.decrement())
print(count1.decrement())


```

```python
class counter:
    def __init__(self,count):
        self.count = count

    def increment(self):
        return self.count + 1

    def decrement(self):
        if self.count > 0:
            return self.count - 1
             
        else:
            print("Count is already zero. Cannot decrement further.")
            
    def reset(self):
        self.count = 0
    

```

```python
count2 = counter(5)
print(count2.count)
print(count2.increment())
print(count2.decrement())
print(count2.reset())
print(count2.increment())
```

```python
class Bird:
    def fly(self):
        print("I can fly!")

class Sparrow(Bird):
    def chirp(self):
        print("Chirp chirp!")

class Penguin(Bird):
    def fly(self):
        print("I cannot fly, but I can swim!")
```

```python

p1 = Sparrow()
p1.fly()

p2 = Penguin()
p2.fly()
```

```python
class Vehicle:
    def __init__(self, brand,wheels):
        self.brand = brand 
        self.wheels = wheels

    def show(self):
        print(f" Brand : {self.brand}  | Wheels {self.wheels}")
            
class Motorcycle(Vehicle):
    def __init__(self,brand,wheels,hhs):
        super().__init__(brand,wheels)
        self.hhs = hhs

    def show(self):
        print(f' Brand : {self.brand}  \n Wheels :{self.wheels}  \n Has_helmet_storage : {self.hhs}')    
```

```python
mo = Motorcycle("BMD",22,"yes")

mo.show()
```
