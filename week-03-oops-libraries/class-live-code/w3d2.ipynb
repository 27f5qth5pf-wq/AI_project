```python
class PYTHON:
    interest_rate = 0.5
    branch = "CS Technology"

    def __init__(self, owner, amount):
        self.owner = owner
        self.balance = amount

    def deposit(self, amount):
        # is_valid = PYTHON.validate_amount(amount)
        if PYTHON.validate_amount(amount):
            self.balance += amount
        else:
            print("Not a valid amount")

    @classmethod
    def change_interest_rate(value):
        PYTHON.interest_rate = value

    @staticmethod
    def validate_amount(amount):
        if amount <= 0:
            return False
        elif amount > 100000:
            return False
        else:
            return True
```

```python
p1 = PYTHON("Ahmad", 1000000)
```
