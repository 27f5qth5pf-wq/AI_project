```python
class LIST(list):
    def __add__ (self,other):
        result = 0
        if self < other:
            for i in self:
                result =self[i] + other[i+1]

        elif self > other:
            for i in other:
                result =other[i] + self[i+1]

        return result
```

```python
class phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
        
    def get_price(self):
        return self.__price

class smartphone(phone):
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera)
        self.os = os
        self.ram = ram
        print("Inside smartphone constructor")

s = smartphone(2000,"samsung",200,'Android',16)

print(s.os)
print(s.get_price())
```

```python
class Sensor:
    def __init__(self, location, readings=None):
        self.location = location
        self.readings = list(readings or [])

    def record(self, value):
        self.readings.append(value)

    def average(self):
        if not self.readings:
            return 0
        return sum(self.readings) / len(self.readings)

    def latest(self):
        if not self.readings:
            return None
        return self.readings[-1]

    def __str__(self):
        return f"Sensor at {self.location} with readings: {self.readings}"


c = Sensor("Room 101", [4, 5, 6])
print("Average:", c.average())
print("Latest:", c.latest())
print("Sensor:", c)
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start(self):
        print(f"{self.make} {self.model} is starting.")

    def stop(self):
        print(f"{self.make} {self.model} is stopping.")


class Car(Vehicle):
    def __init__(self, make, model, year, doors):
        super().__init__(make, model, year)
        self.doors = doors

    def start(self):
        print(f"{self.make} {self.model} with {self.doors} doors is starting.")

    def stop(self):
        print(f"{self.make} {self.model} with {self.doors} doors is stopping.")

class Motorcycle(Car):
    def __init__(self, make, model, year, type):
        super().__init__(make, model, year, doors=0)
        super().__init__(make, model, year)
        self.type = type

    def start(self):
        print(f"{self.make} {self.model} ({self.type}) is starting.")

    def stop(self):
        print(f"{self.make} {self.model} ({self.type}) is stopping.")

motorcycle = Motorcycle("Harley-Davidson", "Street 750", 2021, "Cruiser")
motorcycle.start()
```
